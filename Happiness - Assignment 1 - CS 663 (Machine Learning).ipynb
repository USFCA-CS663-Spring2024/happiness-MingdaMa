### Are there any interesting correlations in the data?
* Generosity has almost-to-none correlation with GDP per capita, it doesn't have a clear correlation with happiness in general as well. However it is very much correlated with people's freedom to make life choices
* Social support has a stronger correlation with happiness than GDP per capita.
* Freedom to make life choices contributes less to happiness than GDP per capita, social support and healthy life expectancy at birth
* GDP per capita has negative correlation with perceptions of corruption. In other words, wealthier countries have less perceptions of corruption than poorer countries

```python
import seaborn as sns

sns.pairplot(
    data=df, 
    y_vars=['Life Ladder'], 
    x_vars=['Log GDP per capita', 
            'Social support', 
            'Healthy life expectancy at birth', 
            'Freedom to make life choices', 
            'Generosity', 
            'Perceptions of corruption', 
            'Positive affect',
            'Negative affect'], 
    kind='scatter')
```

### Can you identify any limitations, missing values or distortions of the data?
* Except for GDP per capita and healthy life expectancy at birth, other data are pulled from polls done by GWP. The poll is based on people's personal opinions, which can be subjective.
* Authoritarian regimes often lack data on perceptions of corruption

```python
poc_null_counts = df.groupby('Country name')['Perceptions of corruption'].apply(lambda x: x.isnull().sum()).reset_index(name='Null Count').sort_values(by='Null Count', ascending=False)
poc_null_counts[poc_null_counts['Null Count'] > 2]
```

# Question 3

### Of the countries in this dataset, which country had the largest increase in happiness from its start of participation to 2023? Which had the largest decrease in happiness?
* Congo (Brazzaville)	 has the largest increase in happiness
* Lebanon has the largest decrease in happiness

```python
df.sort_values(by='year').groupby('Country name')['Life Ladder'].agg(oldest='first', newest='last').assign(life_ladder_difference=lambda x: x['newest'] - x['oldest']).reset_index().drop(['oldest', 'newest'], axis=1).sort_values(by='life_ladder_difference', ascending=False).head()
```
